```python
#2
import pandas as pd
df = pd.read_csv('/Users/kentdong/Desktop/2024Fall/DS2002/acc_players-2324F.csv')
df['PTS'] = pd.to_numeric(df['Unnamed: 13'], errors='coerce')
df['MP'] = pd.to_numeric(df['Unnamed: 6'], errors='coerce')
df['TRB'] = pd.to_numeric(df['Unnamed: 7'], errors='coerce')

total_points = df['PTS'].sum()
print(f"Total Points Scored by All Players Combined: {total_points}")

most_minutes_player = df.loc[df['MP'].idxmax(), ['Unnamed: 1', 'MP']]
print(f"Player with Most Minutes: {most_minutes_player['Unnamed: 1']} with {most_minutes_player['MP']} minutes")

top_5_rebounds = df[['Unnamed: 1', 'TRB']].nlargest(5, 'TRB')
print("Top 5 Players in Terms of Total Rebounds:")
print(top_5_rebounds)
```

```python
#3
import pandas as pd

df = pd.read_csv('/Users/kentdong/Desktop/2024Fall/DS2002/acc_players-2324F.csv')

df['PTS'] = pd.to_numeric(df['Unnamed: 13'], errors='coerce')
df['MP'] = pd.to_numeric(df['Unnamed: 6'], errors='coerce')
df['TRB'] = pd.to_numeric(df['Unnamed: 7'], errors='coerce')
df['AST'] = pd.to_numeric(df['Unnamed: 8'], errors='coerce')
df['BLK'] = pd.to_numeric(df['Unnamed: 9'], errors='coerce')

filtered_df = df[df['MP'] > 500]

highest_assists_player = filtered_df.loc[filtered_df['AST'].idxmax(), ['Unnamed: 1', 'AST']]
print(f"Player with Highest Assists: {highest_assists_player['Unnamed: 1']} with {highest_assists_player['AST']} assists")

top_3_assist_leaders = filtered_df[['Unnamed: 1', 'AST']].nlargest(3, 'AST')
print("Top 3 Assist Leaders:")
print(top_3_assist_leaders)

top_3_shot_blockers = filtered_df[['Unnamed: 1', 'BLK']].nlargest(3, 'BLK')
print("Top 3 Shot Blockers:")
print(top_3_shot_blockers)
```

```python
#4
import pandas as pd

df = pd.read_csv('/Users/kentdong/Desktop/2024Fall/DS2002/acc_players-2324F.csv')

df['PTS'] = pd.to_numeric(df['Unnamed: 13'], errors='coerce')
df['MP'] = pd.to_numeric(df['Unnamed: 6'], errors='coerce')
df['TRB'] = pd.to_numeric(df['Unnamed: 7'], errors='coerce')
df['AST'] = pd.to_numeric(df['Unnamed: 8'], errors='coerce')
df['BLK'] = pd.to_numeric(df['Unnamed: 9'], errors='coerce')

school_points = df.groupby('Unnamed: 4')['PTS'].sum()
print("Total Points Scored by Each School:")
print(school_points)

school_assists = df.groupby('Unnamed: 4')['AST'].sum()
print("\nTotal Assists by Each School:")
print(school_assists)

top_3_schools_by_points = school_points.nlargest(3)
print("\nTop 3 Schools by Total Points Scored:")
print(top_3_schools_by_points)
```

```python
#5
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/Users/kentdong/Desktop/2024Fall/DS2002/acc_players-2324F.csv')

df['PTS'] = pd.to_numeric(df['Unnamed: 13'], errors='coerce')
df['MP'] = pd.to_numeric(df['Unnamed: 6'], errors='coerce')
df['TRB'] = pd.to_numeric(df['Unnamed: 7'], errors='coerce')
df['AST'] = pd.to_numeric(df['Unnamed: 8'], errors='coerce')
df['BLK'] = pd.to_numeric(df['Unnamed: 9'], errors='coerce')
df['FG%'] = pd.to_numeric(df['Shooting'], errors='coerce')
df['WS'] = pd.to_numeric(df['Unnamed: 19'], errors='coerce')

top_5_players_by_points = df[['Unnamed: 1', 'PTS']].nlargest(5, 'PTS')

plt.figure(figsize=(10, 6))
plt.bar(top_5_players_by_points['Unnamed: 1'], top_5_players_by_points['PTS'], color='skyblue')
plt.title('Top 5 Players by Total Points Scored')
plt.xlabel('Player')
plt.ylabel('Total Points Scored')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(df['FG%'], df['WS'], color='blue', alpha=0.5)
plt.title('Correlation between FG% and Win Shares (WS)')
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.grid(True)
plt.show()

correlation_coefficient = df['FG%'].corr(df['WS'])
print(f"Correlation Coefficient between FG% and WS: {correlation_coefficient}")
```
